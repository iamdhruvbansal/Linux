Question 01: What is Linux?
```
  - Linux is a **command line interface based operating system**. Eg: Redhat, CentOS, AWS Linux etc
  - Linux is an **operating system** like Windows & macOS.
  - That help us to **manage computer hardware & resources** & allows us to run our applications.
```

Question 02: What is the difference between hard link & soft link?
```
  - **Hard Link:** It's essentially a **second name or pointer** to the **same exact data** on the hard drive. If we delete the original file, the link still works because the data stays until all links are deleted. Think of it as a clone; they share the same data block.
  - **Soft Link (Symbolic Link):** It's a **shortcut** or a **pointer to the original file's name/path**. If you delete the original file, the soft link breaks and becomes useless, pointing to a non-existent file. It's like a Windows desktop shortcut.
```
-----

Question 03: What is a Kernel in Linux?
```
  - The **Kernel** is the **core** or **heart of the operating system**. It's the first program loaded when the system starts.
  - It acts as the **middleman** (or translator) between the **hardware** (CPU, memory, devices) and the **running software/applications**, managing all resources efficiently.
```
